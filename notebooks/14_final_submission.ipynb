# Production Inference Integrity

A major focus of the final pipeline was ensuring strict training–inference consistency.

## Model-Specific Preprocessing

Each model family used its own fitted preprocessing pipeline loaded directly from its corresponding training run directory:

```text
run_dir / preprocessor.pkl
```

This was critical because:
- Gradient boosting models used unscaled engineered features,
- Logistic Regression and TabNet required fitted `StandardScaler` transformations.

The pipeline correctly applied:
- `.transform()` only,
- never `.fit_transform()` on test data,
- preventing data leakage and scaling inconsistencies.

This eliminated one of the highest-risk deployment failure modes in tabular ML systems.

---

# Fold-Based Prediction Strategy

For each model family:
- Five cross-validation fold models were loaded,
- Predictions were generated independently,
- Fold outputs were averaged into a single robust prediction vector.

Special handling was implemented for TabNet:
- Fold models were stored as `.zip` artifacts,
- Loaded via `model.load_model(path_without_extension)`,
- Avoiding silent corruption associated with improper deserialization.

All fold models loaded successfully and produced stable probability outputs.

---
